```python
%sql
CREATE OR REPLACE VIEW gld_maz_logistics_warehouse.maz_stock_vw_plus
AS
select * from (
select
  inventory.country_code as country_code,
  inventory.product_code as product_code,
  inventory.product_short_code as product_short_code,
  materials.product_name as material_description,
  inventory.plant_code as plant_code,
  materials.plant_name as plant_name,
  inventory.storage_location_code as storage_location_code,
  inventory.batch_code as batch_code,
  materials.sales_uom_code as sales_uom_code,
  materials.product_type_code as material_type,
  materials.product_group as material_group,
  materials.storage_location_name as description,
  materials.storage_location_code as sl,
  materials.currency as currency,
  materials.sales_organization_code as sales_region_code,
  materials.company_code as company_code,
  inventory.unitary_price as unitary_price,
  inventory.unrestricted as unrestricted,
  inventory.unrestricted_hl as unrestricted_hl,
  inventory.unrestricted_ca as unrestricted_ca,
  inventory.unrestricted_pal as unrestricted_pal,
  inventory.value_unrestricted as value_unrestricted,
  inventory.blocked as blocked,
  inventory.blocked_hl as blocked_hl,
  inventory.blocked_ca as blocked_ca,
  inventory.blocked_pal as blocked_pal,
  inventory.value_blocked_stock as value_blocked_stock,
  inventory.in_quality_inspection as in_quality_inspection,
  inventory.in_quality_inspection_hl as in_quality_inspection_hl,
  inventory.in_quality_inspection_ca as in_quality_inspection_ca,
  inventory.in_quality_inspection_pal as in_quality_inspection_pal,
  inventory.value_in_quality_inspection as value_in_quality_inspection,
  inventory.transit_transf as transit_transf,
  inventory.transit_transf_hl as transit_transf_hl,
  inventory.transit_transf_ca as transit_transf_ca,
  inventory.transit_transf_pal as transit_transf_pal,
  inventory.val_in_trans as val_in_trans,
  inventory.returns as returns,
  inventory.returns_hl as returns_hl,
  inventory.returns_ca as returns_ca,
  inventory.returns_pal as returns_pal,
  inventory.value_rets_blocked as value_rets_blocked,
  inventory.restricted_use as restricted_use,
  inventory.restricted_use_hl as restricted_use_hl,
  inventory.restricted_use_ca as restricted_use_ca,
  inventory.restricted_use_pal as restricted_use_pal,
  inventory.value_restricted as value_restricted,
  inventory.val_grbs as val_grbs,
  inventory.val_grbs_hl as val_grbs_hl,
  inventory.val_grbs_ca as val_grbs_ca,
  inventory.val_grbs_pal as val_grbs_pal,
  inventory.total_value as total_value,
  inventory.stock_in_transit as stock_in_transit,
  inventory.stock_in_transit_hl as stock_in_transit_hl,
  inventory.stock_in_transit_ca as stock_in_transit_ca,
  inventory.stock_in_transit_pal as stock_in_transit_pal,
  inventory.value_in_transit as value_in_transit,
  inventory.stock_in_trans as stock_in_trans,
  inventory.stock_in_trans_hl as stock_in_trans_hl,
  inventory.stock_in_trans_ca as stock_in_trans_ca,
  inventory.stock_in_trans_pal as stock_in_trans_pal,
  inventory.value_in_stock_tfr as value_in_stock_tfr,
  inventory.tied_empties as tied_empties,
  inventory.tied_empties_hl as tied_empties_hl,
  inventory.tied_empties_ca as tied_empties_ca,
  inventory.tied_empties_pal as tied_empties_pal,
  inventory.value_tied_empties as value_tied_empties,
  inventory.stock_insertion_date as stock_insertion_date,
  inventory.stock_update_date as stock_update_date
from
  gld_maz_logistics_warehouse.maz_stock_inventory_plus inventory
 left join
  (
    select batch_code, product_short_code, plant_code, storage_location_code, product_name, plant_name, sales_uom_code, product_type_code, product_group, storage_location_name, currency, sales_organization_code, company_code
    from gld_maz_masterdata_materials.maz_materials_plus
    group by batch_code, product_short_code, plant_code, storage_location_code, product_name, plant_name, sales_uom_code, product_type_code, product_group, storage_location_name, currency, sales_organization_code, company_code
  ) materials 
  ON trim(inventory.product_short_code) = trim(materials.product_short_code)
  and trim(inventory.plant_code) = trim(materials.plant_code)
  and trim(inventory.storage_location_code) = trim(materials.storage_location_code)
  and inventory.batch_code = materials.batch_code
  and materials.product_name is not null
  ) where product_short_code is not null
  
```
